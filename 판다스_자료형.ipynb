```python
msg_list = ["numpy", "pandas", "matplotlib"]
data = [] # 초기화

for msg in msg_list:
  record = [msg]
  for x in ["a", "p", "n"]:
    record.append(msg.count(x))
  data.append(record)
data = pd.DataFrame(data, columns=["message", "a", "p", "n"])
display(data)
```
