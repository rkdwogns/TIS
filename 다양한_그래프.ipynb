```python
# 연습: 산점도 그리기
# 평균 10, 분산이 5인 난수(난수의 개수 100000) 배열을 2개 생성하여 각각을 x좌표와 y좌표 목록으로 가지는 산점도를 그리되,
# 점들이 너무 뭉치지 않도록 배려하시오.
x = np.random.normal(10, 5, 100000) # 평균, 분산, 난수개수
y = np.random.normal(10, 5, 100000)
plt.scatter(x, y, alpha=0.03)
plt.title(f'평균 10 분산 5')
```

```python
# 예제에 사용할 데이터
import pandas as pd
data = pd.DataFrame({'수학':[100, 80, 90], '영어':[70, 90, 80]}, index = ['철수', '영희', '민수'])
display(data)
```

```python
# 히트맵 그리기
import seaborn as sns
sns.heatmap(data)
```

```python
sns.heatmap(data, annot = True, fmt= ".2f")
```

```python
sns.heatmap(data, annot = True, fmt= ".2f", 
            cmap="Greys", linewidth=1,
            linecolor= "black")
```
