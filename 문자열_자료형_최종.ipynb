```python
import pandas as pd
path = '/content/drive/MyDrive/Colab Notebooks/주가리포트_샘플.txt'
df = pd.read_csv(path, sep='\t', encoding="euc-kr")
display(df.head())
```

```python
df.info()
```

```python
display(df['의견'].str.contains('매수'))
```

```python
# 연습문제
# '매도'가 포함된 리포트만 필터링해 보세요.
display(df['의견'].str.contains('매도'))
# df['의견'].str.contains('매수')의 결과가 False인 종목의 행을 추출해 보세요.
display(df[df['의견'].str.contains('매수')==False])
```

```python
# '의견' 칼럼에서 'Neutral'이라는 문자를 포함한 행들을 추출하시오.
display(df[df['의견'].str.contains('Neutral')])
```

```python
df['의견'] = df['의견'].str.replace('Neutral', '중립') # inplace 메서드 아님
```
