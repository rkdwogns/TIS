```python
# 매출액, 영업이익에 대해서 각각 최대값과 최소값을 구하시오.
print('매출액 최대값:',df['매출액'].max(), '매출액 최소값:',df['매출액'].min())
print('영업이익 최대값:',df['영업이익'].max(), '영업이익 최소값:',df['영업이익'].min())
```

```python
# 매출액, 영업이익에 대해서 각각 최대값과 최소값인 기업 및 해당연도를 구하시오.
display(df[df['매출액']==df['매출액'].max()])
display(df[df['매출액']==df['매출액'].min()])
display(df[df['영업이익']==df['영업이익'].max()])
display(df[df['영업이익']==df['영업이익'].min()])
```

```python
display(df.groupby('기업')['영업이익'].max())
```

```python
result = df.groupby(['연도', '부채비율>50'])['영업이익'].max()
display(result)
```

```python
display(df.groupby(["연도", "부채비율>50"], as_index = False)["영업이익"].max())
```

```python
# 연습
# 기업별 (5개년도의) 매출액 평균, 최대값, 최소값을 구해 보세요.
display(df.groupby('기업')['매출액'].mean())
display(df.groupby('기업')['매출액'].max())
display(df.groupby('기업')['매출액'].min())
```

```python
# 기업별, 부채비율 50% 초과유무별, 부채 평균을 구하시오.
display(df.groupby(['기업', '부채비율>50'], as_index = False)['부채총계'].mean())
```

```python
import numpy as np
result = df.groupby("기업", as_index = False)[["영업이익", "매출액"]].agg([max, min, np.mean])
display(result)
```

```python
import numpy as np
result = df.groupby("기업", as_index = False)[["영업이익", "매출액"]].agg({'영업이익':[max, min], '매출액':[np.mean]})
display(result)
```
